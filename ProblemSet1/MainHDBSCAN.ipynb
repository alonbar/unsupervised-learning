```python
import numpy             as np
import matplotlib.pyplot as plt
import matplotlib.cm
import matplotlib

matplotlib.rc('font', **{'size' : 16})
```

```python
mX = np.load('clusterable_data.npy')

plt.figure(figsize=(8, 8))
plt.scatter(mX[:,0], mX[:,1], s=50, color=[0,1,0], edgecolor='k')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.grid()
plt.axis('equal')
plt.show()
```

```python
from hdbscan import HDBSCAN

oHDBSCAN = HDBSCAN(min_samples=5)
vC       = oHDBSCAN.fit_predict(mX)
np.unique(vC)
oHDBSCAN
```

```python
Nc = len(np.unique(vC))

plt.figure(figsize=(9, 9))
plt.scatter(mX[:,0], mX[:,1], s=50, c=vC, edgecolor='k', cmap=matplotlib.cm.get_cmap('gist_rainbow', Nc))
plt.axis('equal')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.show()
```
